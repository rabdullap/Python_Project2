```python
##2.Segregate the employees w.r.t different positions.
```

```python
Employee=df.groupby('Position')['Name'].apply (list)
for position, names in Employee.items():
    print(f"Employee in {position} position: ")
    for name in names:
        print(name)
    print("\n")
```

```python
##3.Find from which age group most of the employees belong
```

```python
bins=[ 15,20.25,30,35,40,45]
age=pd.cut(df['Age'],bins=bins)
age
```

```python
Age_group=age.value_counts().idxmax()
print("Highst Number of Employees is From the Age Group ",Age_group)
```

```python
##4Find out under which team and position, spending in terms of salary is high
```

```python
Highst_Salary= df.groupby(["Team","Position",]
                          ,as_index=False)['Salary'].sum().sort_values(by="Salary",ascending=False)
Highst_Salary
```

```python
HS=Highst_Salary ['Salary'].max('index')
HS
```

```python
High_Salary=Highst_Salary.loc[Highst_Salary['Salary']==HS,["Team","Position","Salary",]]
High_Salary
```

```python
##5.Find if there is any correlation between age and salary , represent it visually.
```

```python
x='Age'
y='Salary'
plt.scatter(data["Age"],data["Salary"],color="green")
plt.xlabel(x)
plt.ylabel(y)
plt.title('Correlation between age and salary')
plt.show()
```

```python
c=data[['Age','Salary']].corr()
c     
```

```python
sns.heatmap(c,annot=True,cmap='viridis')
plt.title('Correlation between age and salary')
plt.show()
```
